```python
#K-,eans clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from re import X
from scipy.io import loadmat
mat = loadmat("/content/ex7data2.mat")
X = mat["X"]

#initialise centroids
import random
init_centroids=random.sample(range(0,len(X)),3)
init_centroids
```

```python
centroids = []
for i in init_centroids:
  centroids.append(X[i])
centroids
```

```python
def cal_dist(x1,x2):
  return (sum((x1-x2)**2))**0.5
cal_dist(centroids[1],centroids[2])
```

```python
def findclosestcentroid(centroids, x):
  assigned_centroids = []
  for i in X:
    distance = []
    for j in centroids:
      distance.append(cal_dist(i,j))
    assigned_centroids.append(np.argmin(distance)) # here we use argmin instead of min then because we want number of distnce near centroid
  return assigned_centroids

findclosestcentroid(centroids,X)
```

```python
#mean of the observation falling in each cluster category
def compute_new_centroids(cluster_labels, X):
    new_centers = []
    new_df = pd.concat([pd.DataFrame(X), pd.DataFrame(cluster_labels, columns=["cluster"])], axis=1)
    for c in set(new_df["cluster"]):
        current_cluster = new_df[new_df["cluster"] == c][new_df.columns[:-1]]
        new_mean = current_cluster.mean(axis=0)
        new_centers.append(new_mean)
    return new_centers
```

```python
for i in range(10):
    obs_centroid = findclosestcentroid(centroids, X)
    updated_centroids = compute_new_centroids(obs_centroid, X)
    centroids = updated_centroids

    plt.figure()
    plt.scatter(np.array(centroids)[:, 0], np.array(centroids)[:, 1], color='black')
    plt.scatter(X[:, 0], X[:, 1])
    plt.show()

```
