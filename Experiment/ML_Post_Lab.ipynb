```python
data=pd.read_csv('/content/survey_results_public.csv')
hobby_counts = data['Hobby'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(hobby_counts, labels=hobby_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Do People Find Coding a Hobby?')
plt.axis('equal')
plt.show()
```

```python
open_source_counts = data['OpenSource'].value_counts()
print("\nContributions to Open Source:\n", open_source_counts)
```

```python
top_countries = data['Country'].value_counts().head(20)
plt.figure(figsize=(12,6))
top_countries.plot(kind='bar')
plt.title('Top 20 Countries by Number of Responses')
plt.ylabel('Number of Responses')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#6 particular courses number of students pursue
course_counts = data['UndergradMajor'].value_counts(dropna=False)

# Bar chart visualization
plt.figure(figsize=(10,6))
course_counts.plot(kind='bar')
plt.title('Courses Pursued by Students')
plt.xlabel('Course')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#7 students or working professional are full time or part time
full_time_counts = data['Employment'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(full_time_counts, labels=full_time_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(' Employment Chart')
plt.axis('equal')
plt.show()
```

```python
#8 A company's maximum number of employees
company_size_counts = data['CompanySize'].value_counts()
plt.figure(figsize=(10,6))
company_size_counts.plot(kind='bar')
plt.title('Number of Respondents by Company Size')
plt.xlabel('Company Size Range')
plt.ylabel('Number of Employees (Based on Respondents)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
#9Which Type of Developer a student or employee is?
data = pd.read_csv('/content/survey_results_public.csv')

# Drop NaNs from 'DevType' and split multiple selections into individual rows
dev_types = data['DevType'].dropna().str.split(';')
all_dev_types = [dev.strip() for sublist in dev_types for dev in sublist]
developer_type_counts = pd.Series(all_dev_types).value_counts()

# Plotting
plt.figure(figsize=(12, 8))
developer_type_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Distribution of Developer Types', fontsize=16)
plt.xlabel('Developer Type', fontsize=14)
plt.ylabel('Number of Respondents', fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()  # Avoid label cutoff
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#10 10 countries which have higher number of full time employee
full_time = data[data['Employment'] == 'Employed full-time']
full_time_by_country = full_time['Country'].value_counts()
plt.figure(figsize=(12, 6))
full_time_by_country.head(10).plot(kind='bar', color='teal', edgecolor='black')
plt.title('Top 10 Countries by Number of Full-Time Employees', fontsize=15)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Full-Time Respondents', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
