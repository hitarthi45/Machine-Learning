```python
path = r'/content/exampe_image.jpeg'
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
import numpy as np
from scipy.ndimage import convolve
```

```python
from google.colab.patches import cv2_imshow
img = cv2.imread(path)
cv2_imshow(img)
```

```python
img = cv2.imread(path,cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
#KERNEL DECLARATION
identity=np.array([[0,0,0],[0,1,0],[0,0,0]])
print("Identity\n",identity)
# blur = np.ones((7,7),dtype=float)*(1/49)
blur = np.ones((7,7),dtype=float)*(1/10)
print("Blur\n",blur)
horizontal = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
print("Horizontal\n",horizontal)
```

```python
kernel = horizontal
img2 = convolve(img, kernel)
cv2_imshow(img2)

#image is getting blurred if we increase the value to be divided in the kernel matrix and the image is getting darker
```
